```
# Importanodo pacotes necessÃ¡rios
!pip install pandas
!pip install yfinance
!pip install matplotlib

import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```
# Definindo empresas a serem consultadas
tickers_list = ['KNCR11.SA', 'WEGE3.SA', 'ITSA3.SA', 'KNRI11.SA', 'LREN3.SA', 'WEGE3.SA']
#Definindo DataFrame
data = pd.DataFrame(columns=tickers_list)
print(data)
```

```
# Percorre cada ticker a ser consultado
for ticker in tickers_list:
 # consulta dados on yahoo finance e armazena na coluna com o nome do ticker
 data[ticker] = yf.download(ticker,start="2019-10-17", end="2019-12-31")['Adj Close']
 # Print first 5 rows of the data
data.dropna()  
print(data)
```

```
# Calcula variaÃ§Ã£o acumulada
dataVar = ((data.pct_change()+1).cumprod())
#dataVar = dataVar.fillna(0)

# plota(cria) o grÃ¡fico
dataVar.plot(figsize=(15, 7))
#Exibe a legenda 
plt.legend()
# Define label para o tÃ­tulo do grÃ¡fico
plt.title('VariaÃ§Ã£o do PreÃ§o Ajustado', fontsize=16)
# Define label para os eixos x e y
plt.ylabel('PreÃ§o', fontsize=14)
plt.xlabel('Ano', fontsize=14)
# EstitilizaÃ§Ã£o de linhas
plt.grid(which='major', color='k', linestyle='-.', linewidth=0.5)
# Exibe o grÃ¡fico
plt.show()
```
