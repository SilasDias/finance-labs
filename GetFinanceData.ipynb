```
# Importanodo pacotes necessários
!pip install pandas
!pip install yfinance
!pip install matplotlib

import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```
# Definindo empresas a serem consultadas
tickers_list = ['BRML3.SA', 'CCRO3.SA', 'CGAS5.SA', 'CMIG4.SA', 'LREN3.SA', 'RENT3.SA']
#Definindo DataFrame
data = pd.DataFrame(columns=tickers_list)
print(data)
```

```
# Percorre cada ticker a ser consultado
for ticker in tickers_list:
 # consulta dados on yahoo finance e armazena na coluna com o nome do ticker
 data[ticker] = yf.download(ticker,start="2020-01-31", end="2020-02-06")['Adj Close']
 # Print first 5 rows of the data
data.dropna()  
print(data)
```

```
# Calcula variação acumulada
dataVar = ((data.pct_change()+1).cumprod())
#dataVar = dataVar.fillna(0)

# plota(cria) o gráfico
dataVar.plot(figsize=(15, 7))
#Exibe a legenda 
plt.legend()
# Define label para o título do gráfico
plt.title('Variação do Preço Ajustado', fontsize=16)
# Define label para os eixos x e y
plt.ylabel('Preço', fontsize=14)
plt.xlabel('Ano', fontsize=14)
# Estitilização de linhas
plt.grid(which='major', color='k', linestyle='-.', linewidth=0.5)
# Exibe o gráfico
plt.show()
```
